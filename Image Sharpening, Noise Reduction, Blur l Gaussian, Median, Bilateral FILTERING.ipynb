```python
import cv2
import numpy as np

img = cv2.imread('data/nature.jpg')

#kernal bluring using filter2D
kernal_25 = np.ones((25, 25), np.float32)/625
output_kernal = cv2.filter2D(img, -1, kernal_25)

#blur function blur
output_blur = cv2.blur(img, (25, 25))

#box filter blur
#by using this function we can do all the above step using this fucntion
output_box = cv2.boxFilter(img, -1, (25, 25), normalize=True)

#gaussian bluring better than avg bluring
#Gaussian blur will give high weight to center, surrounding pixels and give law weight to far pixels
# output_gaus = cv2.GaussianBlur(img, ksize, sigmaX)
# sigmaX - Normalization rate
output_gaus = cv2.GaussianBlur(img, (25, 25), 0)

#Median blur
#media blur replace all the pixel value with the median of all value lying in kernal area
#basically used for the reduction of Noise
output_med = cv2.medianBlur(img, 5)

#Bilateral filtering(Reduction of noise + Preserving of edgees)
#Bilateral function make sure that only those pixels having intensity almost same as target pixel are considered.
# output_bil = cv2.bilateralFilter(src, d, sigmaColor, sigmaSpace)
# src - screen
# d - distance consider
# sigmaColor - if i give 6 the closest 6 color combined and create new color from it 
# sigmaSpace
output_bil = cv2.bilateralFilter(img, 5, 6, 6)



cv2.imshow('orginal', img)
cv2.imshow('filter2D blur', output_kernal)
cv2.imshow('blur function', output_blur)
cv2.imshow('box filter', output_box)
cv2.imshow('GaussianBlur', output_gaus)
cv2.imshow('medianBlur', output_med)
cv2.imshow('Bilateral', output_bil)



cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Image Sharpening
Basically the difference between orginal image and blured image<br>


```python
import cv2
import numpy as np

img = cv2.imread('data/nature.jpg')

#Gaussian kernal for sharpening
gaussian_blur = cv2.GaussianBlur(img, (5,5), 0)

#Sharpening using addWeighted()
# Sharpening = cv2.addWeighted(src1, alpha, src2, beta, gamma)
# alpha(1.5) - beta(0.5) need to be 1
# gamma intensity
Sharpening1 = cv2.addWeighted(img, 1.5, gaussian_blur, -0.5, 0)
Sharpening2 = cv2.addWeighted(img, 3.5, gaussian_blur, -2.5, 0)
Sharpening3 = cv2.addWeighted(img, 7.5, gaussian_blur, -6.5, 0)

cv2.imshow('Sharpening1', Sharpening1)
cv2.imshow('Sharpening2', Sharpening2)
cv2.imshow('Sharpening3', Sharpening3)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
