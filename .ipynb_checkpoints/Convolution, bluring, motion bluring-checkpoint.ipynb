# Convolution
convolution is a fundamental operation in image processing. We basically a mathematical operator to each pixel and change its value in some way.To applay this mathematical operators we use another metrix called kernel

It the kernal is out of image metix then openCV will automatically give value zero to those matrix or replace with neighbor

## IMP
###### Why are kernel's dimensions odd ?


### Computer Vision Terminology<br>
The kernal is called the "image filter" <br> and the process of applying this kernal to the given image is called "image filtering"<br>
The output obtained after applying the kernal to the image is called "filter image"

### What is frequency in image ?
###### Frequency refers to the   rate of change of pixel values. So we can say that the sharp edges woulb be high frequency content because the pixel value changes rapidly in the region. Going  by that logic, plain area would be law frequency content.<br>


# Bluring

```python
# with out normalizing the filters
import cv2
import numpy as np

img = cv2.imread('data/nature.jpg')

kernal_identity = np.array([[0, 0, 0], [0, 1, 0], [0, 0, 0]])
kernal_3 = np.ones((3, 3), dtype = np.float32)
kernal_11 = np.ones((11, 11), dtype = np.float32)

output_1 = cv2.filter2D(img, -1, kernal_identity)
output_2 = cv2.filter2D(img, -1, kernal_3)
output_3 = cv2.filter2D(img, -1, kernal_11)

cv2.imshow('orginal',img)
cv2.imshow('kernal_3', output_2)
cv2.imshow('kernal_11', output_3)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# with normalizing the filters
import cv2
import numpy as np

img = cv2.imread('data/nature.jpg')
# 3 * 3 = 9
kernal_identity = np.array([[0, 0, 0], [0, 1, 0], [0, 0, 0]])/9
# 3 * 3 = 9
kernal_3 = np.ones((3, 3), dtype = np.float32)/9
# 11 * 11 = 121
kernal_11 = np.ones((11, 11), dtype = np.float32)/121

output_1 = cv2.filter2D(img, -1, kernal_identity)
output_2 = cv2.filter2D(img, -1, kernal_3)
output_3 = cv2.filter2D(img, -1, kernal_11)

cv2.imshow('orginal',img)
cv2.imshow('kernal_3', output_2)
cv2.imshow('kernal_11', output_3)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Motion Bluring
In motion bluring we are averaging out the values in a perticular direction

```python
import cv2
import numpy as np

img = cv2.imread('data/nature.jpg')
size = 15
kernel = np.zeros((size, size))

# Fill the middle row with ones
kernel[:, int((size-1)/2)] = np.ones(size)
# here we normalize kernal only with one size
kernel = kernel/size

output = cv2.filter2D(img, -1, kernel)
cv2.imshow('motionblur', output)
cv2.imshow('orginal', img)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

img = cv2.imread('data/nature.jpg')
size = 15
kernel = np.zeros((size, size))

# Fill the middle row with ones
kernel[int((size-1)/2),:] = np.ones(size)
kernel = kernel/size

output = cv2.filter2D(img, -1, kernel)
cv2.imshow('motionblur', output)
cv2.imshow('orginal', img)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
