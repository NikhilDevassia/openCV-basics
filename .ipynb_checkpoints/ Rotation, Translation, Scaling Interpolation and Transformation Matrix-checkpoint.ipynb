# Linear interpolation between two linear points

#### It is the process of estimating the values of a continuous fuction from discrete samples.

<div>

</div>
There are basically 5 types of interpolations in opencv
<br>
<br>
1)Linear interpolation<br>
2)Area interpolation<br>
3)Cubic interpolation<br>
4)Nearest neighbor interpolation<br>
5)Sinusoidal interpolation<br>

```python
import cv2
import numpy as np


image = cv2.imread('data/nature.jpg')

image_sized = cv2.resize(image, (300,300))

#Resizing image using Linear interpolation
image_re_linear = cv2.resize(image, None, fx=5.5,fy=5.5, interpolation=cv2.INTER_LINEAR)

#Resizing using Cubic interpolation
image_re_cubic = cv2.resize(image, None, fx=5.5,fy=5.5, interpolation=cv2.INTER_CUBIC)

#Showing all three images
cv2.imshow('Linear', image_re_linear)
cv2.imshow('Cbubic', image_re_cubic)
cv2.imshow('original', image)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Translation
###### Image translation means shifting image in co-ordinate space by adding a special value to x and y co-ordinate<br>

T = [1 , 0, tx],<br>
......[0, 1, ty ]

```python
import cv2
import numpy

img = cv2.imread('data/nature.jpg')

#Translation metrix
matrix = numpy.float32([[1, 0, 100], [0, 1, 100]])

#Applaying the metrix to the image
#if translate image like this, then the image will be croped so we add extra value to img (+100)
# translate = cv2.warpAffine(img, matrix, (img.shape[1], img.shape[0]))
translate = cv2.warpAffine(img, matrix, (img.shape[1]+100, img.shape[0]+100))


cv2.imshow('translation', translate)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Rotation

```python
import cv2
import numpy 

img = cv2.imread('data/nature.jpg')

#image have three value height, width, channels
height, width = img.shape[:2]
print(height)
print(width)

#Rotation matrix
# matrix = cv2.getRotationMatrix2D(center, angle, scale)
matrix = cv2.getRotationMatrix2D((width/2, height/2), 10, 1)

#Applaying the matrix to 
rotation = cv2.warpAffine(img, matrix,(width, height))

cv2.imshow('rotation', rotation)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
